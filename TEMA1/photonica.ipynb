```python
from analisisb_V import b_v
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Vamos a hacer un código para hacer un diagrama b-V

N_max = 400 # numero de puntos a representar
b = np.linspace(0,1, N_max) # valores de b entre 0 y 1
nus = [0, 1, 2] # valores de nu
aa = [0.0, 50.0] # valores de a
plt.figure()
for nu in nus:
    for a in aa:
        b, V = b_v(nu, a, N_max)
        plt.plot(V, b, label = r'$\nu = %d, a = %.1f$' % (nu, a))
plt.grid()
plt.legend(loc='best')
plt.xlabel('V')
plt.ylabel('b')
plt.xlim(0, 20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N_max = 400  # number of points for plot
b = np.linspace(0, 1.0, N_max)

plt.figure()
for nu in [0, 0.5, 1, 2]:
    for a in [0.0, 8.0, 50.0]:  # asymmetry coefficient
        # determine V
        V1 = np.arctan(np.sqrt(b / (1 - b)))
        V2 = np.arctan(np.sqrt((b + a) / (1.0 - b)))
        V3 = 1 / np.sqrt(1.0 - b)
        V = (nu * np.pi + V1 + V2) * V3

        plt.plot(V, b, linewidth=1.2)

plt.grid(True)
plt.axis([0.0, 20.0, 0.0, 1.0])  # change axis limit
plt.xlabel('V', fontsize=14)
plt.ylabel('b', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.text(10, 0.96, r'$\nu=0$', fontsize=14)
plt.text(10, 0.8, r'$\nu=1$', fontsize=14)
plt.text(10, 0.55, r'$\nu=2$', fontsize=14)

plt.text(0.1, 0.3, 'a=0', fontsize=14)
plt.text(2.5, 0.3, 'a=50', fontsize=14)

plt.text(3.3, 0.2, 'a=0', fontsize=14)
plt.text(5.8, 0.2, 'a=50', fontsize=14)

plt.text(6, 0.1, 'a=0', fontsize=14)
plt.text(8.8, 0.1, 'a=50', fontsize=14)

plt.show()
```
